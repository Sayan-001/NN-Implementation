### Importing preliminary Libraries

```python
from sklearn.datasets import make_regression, make_classification
from sklearn.model_selection import train_test_split
```

### Testing Linear Regression Model

```python
X, y = make_regression(n_samples=5000, n_features=200, n_informative=100, noise=0.3)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from models.Linear_Regression import LinearRegressor

linear_model = LinearRegressor(X.shape[1])
history = linear_model.train(X_train, y_train, X_test, y_test, epochs=25, lr=0.0001, log=True)
linear_model.plot(history)
```

### Testing Logistic Regression Model

```python
X, y = make_classification(n_samples=2000, n_features=500, n_informative=200,  n_classes=2)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from models.Logistic_Regression import LogisticRegressor

model = LogisticRegressor(X.shape[1])
history = model.train(X_train, y_train, X_test, y_test, epochs=100, lr=0.001, log=True)
model.plot(history)
```

### Testing Neural Network

```python
X, y = make_classification(n_samples=5000, n_features=500, n_classes=2)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from models.Neural_Network import Single_Layer_NN

model = Single_Layer_NN(no_of_features=X.shape[1], neurons=64, hidden_layer_activation="relu", output_layer_activation="sigmoid")
history = model.train(X_train, y_train, X_test, y_test, epochs=50, lr=0.001, log=True)
model.plot(history)
```
